<h3>处理数据不平衡</h3>

<pre>
这个数据的一个特点就是不平衡，不论是目标特征还是其他特征，他们绝大部分数值很集中，会有少数数据和其他相差比较大，但又很难认定为异常值


SeriousDlqin2yrs变量0和1的比值超过13,这种数据做test总是会拟合度很高，但没有意义
我就用两个简单的方法处理这个问题，先上采样，但因为容易导致过拟合，所以首先用smote来做上采样，只让0和1比例调整到6：1，之后再做下采样
这个smote的函数是借用简书上一位同学的：http://shataowei.com/2017/12/01/python%E5%BC%80%E5%8F%91%EF%BC%9A%E7%89%B9%E5%BE%81%E5%B7%A5%E7%A8%8B%E4%BB%A3%E7%A0%81%E6%A8%A1%E7%89%88-%E4%B8%80/
就是通过求各个点k临近点的均值（如果是离散变量就直接用众数)，来生成新的数据

用这个函数要注意一个问题，函数根据这个变量的unique值的数量来判断其是连续变量和离散变量，默认是10,也就是说当一个变量有9个不同值，就会把它
当作离散数据，但是默认我们造新的点是取5个临近点，有一定几率会取到5个不同的值，求众数返回的是空值，导致最后又生成了一个带有少量缺失值的数据集
所以这个函数还是要修改的，例如取不到mode时随机选一个值，不过也可以直接把kdistinctvalue改小一些，对于我来说9个不同值的变量也是连续变量

下采样如果直接删掉一部分数据，可能会损失很多信息，所以我选择在用LR之前，随机有放回地抽出一些副样本，生成10组数据，每组里面0和1比例为3：1
用这些数据来进行学习之后再求均值，过程有点像交叉验证



```python
resample_train = smote(train, tag_index=0, std_rate=10, kdistinctvalue=8)
resample_train.columns = ['RevolvingUtilizationOfUnsecuredLines','age','DebtRatio','MonthlyIncome',
                          'NumberOfOpenCreditLinesAndLoans', 'NumberRealEstateLoansOrLines',
                          'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents', 'SeriousDlqin2yrs']


resample_train[['age', 'MonthlyIncome', 'NumberOfOpenCreditLinesAndLoans', 'NumberRealEstateLoansOrLines', 
                'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents']] = 
resample_train[['age', 'MonthlyIncome', 'NumberOfOpenCreditLinesAndLoans', 'NumberRealEstateLoansOrLines', 
                'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents']].apply(round)
```

```python
resample_train['RevolvingUtilizationOfUnsecuredLines'][resample_train['RevolvingUtilizationOfUnsecuredLines']>=1]=2
test['RevolvingUtilizationOfUnsecuredLines'][test['RevolvingUtilizationOfUnsecuredLines']>=1]=2
```
