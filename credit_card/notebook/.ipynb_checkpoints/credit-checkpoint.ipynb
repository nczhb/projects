```python
def cut(x):
    groups = {}
    for i in range(len(x.index.unique())-1):
        y1 = x.ix[:i+1]
        y2 = x.ix[i+2:]
        p1 = float(y1.sum()/len(y1))
        p2 = float(y2.sum()/len(y2))
        p0 = float(x.sum()/len(x))
        p11= len(y1)/len(x)
        p22= len(y2)/len(x)
        part1 = 1-p1**2-(1-p1)**2
        part2 = 1-p2**2-(1-p2)**2
        part0 = 1-p0**2-(1-p0)**2
        p = part0-p11*part1-p22*part2
        groups[i+1] = [part0, part1, part2, p]
    gain_max = max(list(zip(*groups.values()))[3])
    print(gain_max)
    for pos,group in groups.items():
        if group[3]==gain_max:
            print(pos, group)
```
