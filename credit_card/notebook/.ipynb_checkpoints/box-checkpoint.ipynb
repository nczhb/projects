<h3>将连续变量分箱</h3>

```python
resample_train = pd.read_csv('../resample_train1.csv')
```

<pre>
分箱方法我选择用cart，这个是在决策树里的一个算法，通过计算基尼系数来确定从哪个点分组
通常这个方法是用于类型变量的，对于连续变量，我是将它们先排序，然后从不同位置把数据分开，计算相应的基尼系数，挑选变化最大的切分点

```python
def cut(x):
    groups = {}
    index_uni = np.sort(x.index.unique())
    for i in range(len(index_uni)-1):
        index1 = index_uni[:i+1]
        index2 = index_uni[i+1:]
        y1 = x.ix[index1]
        y2 = x.ix[index2]
        p1 = float(y1.sum()/len(y1))
        p2 = float(y2.sum()/len(y2))
        p0 = float(x.sum()/len(x))
        p11= len(y1)/len(x)
        p22= len(y2)/len(x)
        part1 = 1-p1**2-(1-p1)**2
        part2 = 1-p2**2-(1-p2)**2
        part0 = 1-p0**2-(1-p0)**2
        p = part0-p11*part1-p22*part2
        groups[index_uni[i+1]] = [part0, part1, part2, p]
    gain_max = max(list(zip(*groups.values()))[3])
    print('max_gain:', gain_max)
    for pos,group in groups.items():
        if group[3]==gain_max:
            print('cut_position:', pos)
            return pos
```

```python
def spliter(data):
    index_uni = list(np.sort(data.index.unique()))
    index_sets = [index_uni]
    while True:
        if len(index_sets)>=10:break
        #print('index_sets', index_sets)
        index_set_len = list(map(len, index_sets))
        #print('index_set_len:',index_set_len)
        longest_pos =[x for x,y in enumerate(index_set_len) if y==max(index_set_len)]
        #print('longest_pos:', longest_pos)
        longest_sets = [index_sets[pos] for pos in longest_pos]
        for set in longest_sets:
            index_sets.remove(set)
        #print('after delete:', index_sets)
        #print('longest_sets', longest_sets)
        for set in longest_sets:
            #print('set:', set)
            new_pos = cut(data.ix[set])
            #print('new_pos:', new_pos)
            split_pos = set.index(new_pos)
            #print('split_pos:', split_pos)
            part1 = set[:split_pos]
            part2 = set[split_pos:]
            index_sets.append(part1)
            index_sets.append(part2)
            print('after fit in:')
            print(np.sort(index_sets), '\n\n')
```

<pre>
cut函数用于计算各切分点的gini系数
spliter函数每次调用cut选择系数最高的点分割数据，然后对新生成的几组数据里长度最大的一组数据再进行切分，分到10组为止
spliter只是作为一个测试手段，先分出10组，然后根据基尼系数和切分结果(例如各分组包含的数据量)选择需要的切分点
