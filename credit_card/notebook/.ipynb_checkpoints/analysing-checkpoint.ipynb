<pre>
woe值，是对特征分箱之后形成的各个组来计算的，每个组的woe值等于(该组good用户占所有good用户的比值)和(该组bad用户占所有bad用户的比值)相除再取ln

\begin{align}
WOE_i &= ln(\frac{\frac{good_{组内}}{good_总}}{\frac{bad_{组内}}{bad_总}})
\end{align}

<pre>
特征的iv值，就是该特征各组内good用户占比和bad用户占比之差乘该组woe值，然后求和

\begin{align}
IV &= \sum_{i=1}^N{({\frac{good_{组内}}{good_总}} - {\frac{bad_{组内}}{bad_总}})*WOE_i}
\end{align}

```python
def good(x):
    return (len(x)-sum(x))/good_sum
def bad(x):
    return sum(x)/bad_sum

bad_sum = int(resample_train['SeriousDlqin2yrs'].sum())
good_sum = len(resample_train['SeriousDlqin2yrs']) - bad_sum
```

```python
def woe_iv(var, cut):
    data = Series(resample_train['SeriousDlqin2yrs'].values, resample_train[var].values)
    data.index = pd.cut(data.index, cut)
    ratio = data.groupby(data.index).agg([good,bad])
    woe = np.log(ratio['good']/ratio['bad'])
    sub = ratio['good']-ratio['bad']
    iv = (sub*woe).sum()
    ivs[var] = iv
    print('woe:')
    print(woe, '\n')
    print('iv:', iv)
    resample_train[var] = pd.cut(resample_train[var], cut).map(woe.to_dict())
    test[var] = pd.cut(test[var], cut).map(woe.to_dict())
    
```

```python
def undersample(data, subsets=7):
    train_sets = []
    neg = resample_train[resample_train['SeriousDlqin2yrs']==0]
    pos = resample_train[resample_train['SeriousDlqin2yrs']==1]
    for i in range(subsets):
        re_neg = neg.sample(frac=0.5)
        combine = pd.concat([re_neg, pos])
        train_sets.append(combine)
    return train_sets
```

```python
train_sets = undersample(resample_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
```

```python
def train(data_sets):
    test_results = []
    train_results = []
    test_values = []
    coefs = []
    intercepts = []
    scores = []
    train_values = []
    model = LogisticRegression()
    test_data = test[['RevolvingUtilizationOfUnsecuredLines','age','DebtRatio', 
                        'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents']]
    train_data = resample_train[['RevolvingUtilizationOfUnsecuredLines','age','DebtRatio', 
                        'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents']]
    for data in data_sets:
        y_train = data['SeriousDlqin2yrs']
        x_train = data[['RevolvingUtilizationOfUnsecuredLines','age','DebtRatio', 
                        'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents']]
        clf = model.fit(x_train.values, y_train.values)
        scores.append(clf.score(x_train.values, y_train.values))
        coefs.append(clf.coef_)
        intercepts.append(clf.intercept_)
        test_results.append(clf.predict_proba(test_data)[:,1])
        train_results.append(clf.predict_proba(train_data)[:,1])
        test_values.append(clf.predict(test_data))
        train_values.append(clf.predict(train_data))
    return {'test_results':test_results, 'train_results':train_results, 'coefs':coefs, 'intercepts':intercepts,
            'scores':scores, 'test_values':test_values, 'train_values':train_values}
```

```python
final = train(train_sets)
```

```python
from functools import reduce
```

```python
test_result = reduce(np.add, final['test_results'])/10
train_result = reduce(np.add, final['train_results'])/10
coef = reduce(np.add, final['coefs'])/10
intercept = np.mean(final['intercepts'])
score = reduce(np.add, final['scores'])/10
avg = (reduce(np.add, final['test_values'])/10).mean()
```

<code>
进行LR回归后，我们得到各个特征在模型里的系数以及用户是bad用户的概率p

评分卡可以根据用户good或bad概率比的对数值来定义，也就是
score=A-B*ln(θ)
其中：θ=p/(1-p)

接下来就要确定系数A、B
先设定一个基准比例，例如通常情况，good和bad概率比是20：1,那么ratio=1/20
    然后给这个基准情况定一个分数，一般用600
假定ratio*2时，分数变化为多少，一般会给到20、30、50等，我就设定为30

\begin{align*}
600 &= A-Bln(1/20) \\
600-50 &= A-Bln(2/20)
\end{align*}

```python
θ = 1/20
b = 30/np.log(2)
a = 600 + b*np.log(θ)
```

<code>
接下来就可以计算分数了，ln里的(1-p)/p，实际上可以用变量值和LR系数得到
因为根据LR,概率p是由θ.T*x再套sigmoid函数得到的，所以：

\begin{align}
ln(\frac{p}{1-p})=\theta^Tx
\end{align}


<code>
进而score就可以用回归系数β、变量值(也就是相应的woe值)、A、B来表示

\begin{align}
score&=A-B*β_0-B*β_1*w_1-···-B*β_n*w_n
\end{align}

```python
score = \
a - b*intercept -\
-0.60090275*b*test['RevolvingUtilizationOfUnsecuredLines'] -\
-0.26239358*b*test['age'] -\
-0.39983689*b*test['DebtRatio'] -\
-0.51417891*b*test['NumberOfTime60-89DaysPastDueNotWorse'] -\
-0.48478984*b*test['NumberOfDependents']
```
